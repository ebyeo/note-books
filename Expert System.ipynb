Here we use the predicted exchange rates to perform trading.

```python
import pandas as pd
from datetime import datetime

# Load data
dataset = pd.read_csv('predicted_next_week_rates.csv').drop('Unnamed: 0',axis=1)
```

Starting capital is 10000 for each of the 3 currencies

Looping through prediction to decide which currency to buy or sell

```python
import operator
sg_fund = 10000
us_fund = 10000
p_fund = 10000
overall_profit_or_loss = 0
for index, row in dataset.iterrows():
    rank_rates = {}
    #determine which currency will appreciate the most the following week, also minus trading cost of 1 percent
    sg_d_percent_change = (((row['predicted_next_week_SG/D'] - row['SG/D']) / float(row['SG/D'])) * 100) - 1
    rank_rates['SG_D'] = sg_d_percent_change
    p_d_percent_change = (((row['predicted_next_week_P/D'] - row['P/D']) / float(row['P/D']) * 100)) - 1
    rank_rates['Pound_D'] = p_d_percent_change
    sorted_rates = sorted(rank_rates.items(), key=operator.itemgetter(1), reverse=True)
    best_currency = sorted_rates[0][0]
    print "Currency exchange with highest positive predicted percentage increase is {}".format(best_currency)
    #Perform currency exchange on most promising predicted rates
    if best_currency == 'SG_D':
        if us_fund:
            sg_fund = sg_fund + (float(us_fund)/row['SG/D'])
            us_fund = 0
        if p_fund:
            sg_fund = sg_fund + (float(p_fund)/(row['SG/D']/row['P/D']))
            p_fund = 0
    elif best_currency == 'P_D':
        if us_fund:
            p_fund = p_fund + (float(us_fund)/row['P/D'])
            us_fund = 0
        if sg_fund:
            p_fund = p_fund + (float(sg_fund)/(row['P/D']/row['SG/D']))
            sg_fund = 0
    else:
        if sg_fund:
            us_fund = us_fund + (float(sg_fund)*row['SG/D'])
            sg_fund = 0
        if p_fund:
            us_fund = us_fund + (float(p_fund)*row['P/D'])
            p_fund = 0
    if index > 0:
        #Process profit and loss statement
        #First convert all to USD
        if p_fund > 10000:
            excess = p_fund - 10000
            profit_or_loss = excess*row['P/D']
        elif sg_fund > 10000:
            excess = sg_fund - 10000
            profit_or_loss = excess*row['SG/D']
        elif us_fund > 10000:
            profit_or_loss = us_fund - 10000
        overall_profit_or_loss += profit_or_loss
        print "Profit or Loss for end of trading period is USD ${:,.2f}".format(overall_profit_or_loss)
        print "\n"

```

```python
print "Final P&L is USD ${:,.2f}".format(overall_profit_or_loss)
```
